# Tokenization
In this notebook we shown we perform the cleaning of the tweet data with a custom processing and then the tokenization using Roberta tokenizer

```python
import pandas as pd
import re
import pathlib
import ast
import numpy as np
import nltk
import nltk.tokenize as tok
import random
from algorithms import *
from transformers import AutoTokenizer, RobertaModel, RobertaTokenizer, AutoModel

nltk.download('punkt')

DATASET_DIR = pathlib.Path('../data')
pd.options.plotting.backend = "plotly"
```

```python
def test_tokenizer(string, tok):
    """
    Prints the tokenizer output on a given string or a list of strings.

    Parameters:
    - string: The input string or list of strings to be tokenized.
    - tok: The tokenizer object used for tokenization.

    Returns:
    None
    """
    if isinstance(string, list):
        for i in string:
            print('Original string:')
            print(i)
            
            print('From tokenizer:')
            print(tok.decode(tok.encode(i)))

            print('From our tokenizer:')
            print(tok.decode(tokenize(i, tok)))
            print('')
    else:
        print('Original string:')
        print(string)

        print('From orig tokenizer:')
        print(tok.decode(tok.encode(string)))
        
        print('From our tokenizer:')
        print(tok.decode(tokenize(string, tok)))
```

```python
df = pd.read_csv(DATASET_DIR / 'merged_processed_dataset.csv', sep=';', encoding='utf-8')
df['hard_label'] = df['hard_label'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
name = "FacebookAI/xlm-roberta-large"
tokenizer = AutoTokenizer.from_pretrained(name)
#model = AutoModel.from_pretrained(name)
```

```python
test_sentences_2 = []
random.seed(10)
test_sentences_2.extend(list(df.loc[df['language'] == 'es']['tweet'].sample(2).values))
test_sentences_2.extend(list(df.loc[df['language'] == 'en']['tweet'].sample(2).values))

test_tokenizer(test_sentences_2, tokenizer)
```

```python
test_tokenizer('   No te van a condenar por violar a una mujer…no tengas miedo, las violaciones se pueden comprobar, y ya te digo, si eres padre de una niña, te dejará mas tranquilo saber que existe esta ley y no que encima la culpa sea de tu hija que llevaba la falda muy corta….', tokenizer)
```
