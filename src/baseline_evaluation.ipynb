# Evaluation testing
In this notebook we try the library PyEvALL mandatory for the evaluation as written in the challenge.
We use it on the baseline and gold labels provided just to understand the output.
The metric used are ICM, ICMNorm and FMeasure for the gold hard labels and ICMSoft, ICMSoftNorm for the gold soft labels. 

```python
from os import listdir
from os.path import isfile, join
from pyevall.evaluation import PyEvALLEvaluation
from pyevall.utils.utils import PyEvALLUtils
import json
import pathlib

DATASET_DIR = pathlib.Path('../data')
```

```python
def evaluation_prediction(path_predictions, path_golds, metrics=["ICM", "ICMNorm", "FMeasure"]):
    test = PyEvALLEvaluation()
    params= dict()

    TASK_3_HIERARCHY = {"YES":["IDEOLOGICAL-INEQUALITY","STEREOTYPING- DOMINANCE","OBJECTIFICATION", "SEXUAL-VIOLENCE", "MISOGYNY-NON-SEXUAL- VIOLENCE"], "NO":[]}
    params[PyEvALLUtils.PARAM_HIERARCHY]= TASK_3_HIERARCHY
    params[PyEvALLUtils.PARAM_REPORT]= PyEvALLUtils.PARAM_OPTION_REPORT_DATAFRAME  
    # or params[PyEvALLUtils.PARAM_REPORT]= PyEvALLUtils.PARAM_OPTION_REPORT_EMBEDDED
    report= test.evaluate(path_predictions, path_golds, metrics, **params)
    report.print_report()
    
    return report
```

```python
# can be broken, be carefull
def evaluation_predictions(path_predictions, path_golds, metrics):
    lst_pred=[]
    onlyfiles = [f for f in listdir(path_predictions) if isfile(join(path_predictions, f))]
    for file in onlyfiles:
        lst_pred.append(path_predictions + file)
    params= dict()
    prueba = PyEvALLEvaluation()
    TASK_3_HIERARCHY = {"YES":["IDEOLOGICAL-INEQUALITY","STEREOTYPING- DOMINANCE","OBJECTIFICATION", "SEXUAL-VIOLENCE", "MISOGYNY-NON-SEXUAL- VIOLENCE"], "NO":[]}
    params[PyEvALLUtils.PARAM_HIERARCHY]= TASK_3_HIERARCHY
    params[PyEvALLUtils.PARAM_REPORT]= PyEvALLUtils.PARAM_OPTION_REPORT_DATAFRAME
    
    report= prueba.evaluate_lst(lst_pred, path_golds, metrics, **params)
    report.print_report_tsv()
    return report

# Example of usage
"""
PROVA_DIR = pathlib.Path('../test_prova_json')
evaluation_predictions(PROVA_DIR / 'predictions/hard/',
                       PROVA_DIR / 'golds/prova_gold_hard.json', 
                       metrics=["ICM", "ICMNorm", "FMeasure"]);
"""
```
